```python
from igraph import *
import pygraphviz as pgv
import threading

from IPython.display import Image, display, display_png, display_svg, SVG, clear_output
```

```python
def add_vex(num):
    global total_num
    for i in range(num):
        g.add_vertices(1)
        gviz.add_node(total_num)
        total_num += 1

def add_edge(i,j):
    g.add_edges([(i,j)])
    gviz.add_edge((i,j))

def add_edges(list_):
    g.add_edges(list_)
    for edge in list_:
        gviz.add_edge((edge[0],edge[1]))
    
def display_g(g, layout= 'dot', fformat = 'png'):
    if fformat not in ['png','svg']:
        print 'format is not real',fformat
        return
    name ='igraph.'+fformat
    g.draw(name, prog = layout)
    if fformat == 'svg':
        display(SVG(name))
    else:
        display(Image(name))   
```
