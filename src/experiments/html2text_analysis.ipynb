```python
%matplotlib inline

from html2text import html2text
import requests
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
resp = requests.get('http://themindunleashed.com/2015/11/scientists-invent-nanotechnology-that-connects-your-brain-to-your-computer.html')
text = html2text(resp.text)
paras = text.split('\n')
len(paras)
para_lenghts = [len(p) for p in paras]
```

```python
print text
```

```python
fig = plt.figure()
fig.suptitle('paragraph lengths '+ str(len(para_lenghts)), fontsize=20)
plt.xlabel('para', fontsize=16)
plt.ylabel('length', fontsize=16)
plt.rcParams["figure.figsize"] = (30,3)
plt.plot(para_lenghts)
```

look only at thos with less then 100 character

```python
shrinked = [p for p in para_lenghts if p < 100]
fig = plt.figure()
plt.rcParams["figure.figsize"] = (30,3)
# fig.suptitle('test title', fontsize=20)
plt.xlabel('p', fontsize=16)
plt.ylabel('length', fontsize=16)
plt.plot(shrinked)
```

a historgram

```python
hist, bins = np.histogram(shrinked)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width)
plt.show()
```
