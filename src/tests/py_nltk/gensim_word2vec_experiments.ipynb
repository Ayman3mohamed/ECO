```python
countries = []
leftout = []
similarities = {}

with open("data/list_of_all_countries.txt", 'r') as countries_:
    for country in countries_:
        country = country.strip()
        if country in model:
            countries.append(country.strip())
        else:
            leftout.append(country)
for country in countries:
    sim = []
    for other_country in countries:
        similarity = model.similarity(country,other_country)
        if similarity < 0.9999: # cuz of "Spain" , "Spain" similarity. do: similarity - 1.0
            # doesnt help actually
            sim.append((other_country,similarity))
    similarities[country] = sim
```
